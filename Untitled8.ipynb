```python

target_col='target'
num_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_features = df.select_dtypes(include=['object', 'category']).columns.tolist()

if target_col in num_features:
    num_features.remove(target_col)
if target_col in cat_features:
    cat_features.remove(target_col)

# Pearson Correlation (Numerical Features)
if num_features:
    plt.figure(figsize=(10, 6))
    corr_matrix = df[num_features + [target_col]].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Feature Correlation with Target")
    plt.show()
```

```python
from sklearn.feature_selection import f_classif

# ANOVA for Categorical Features (if target is numeric)
if cat_features and df[target_col].dtype in ['int64', 'float64']:
    print("\nANOVA Test Results (Categorical Features → Numeric Target):")
    for col in cat_features:
        encoded_feature = pd.get_dummies(df[col])  # Convert categorical to numerical
        f_stat, p_value = f_classif(encoded_feature, df[target_col])  # Compute ANOVA

        print(f"{col}: F-stat={f_stat[0]:.2f}, p-value={p_value[0]:.4f}")
```

```python
strong_features = ['has_photo', 'is_blacklisted', 'is_verified', 'can_invite_to_group',
                      'has_occupation', 'audio_available', 'can_send_message', 'has_status', 'has_website','avg_keywords','avg_text_length','reposts_ratio']
```

```python
df.isnull().sum()
```

```python
threshold= 0.5*len(df)
df.dropna(thresh=threshold,axis=1,inplace=True)
print(df.isnull().sum())
```

```python
# Define unknown values to replace
unknown_values = ['unknown', '?', 'N/A', 'None', '']

# Replace with NaN
df.replace(unknown_values, np.nan, inplace=True)
```

```python
X=df[strong_features]
y=df['target']
```

```python
categorical_features = X.select_dtypes(include=['object']).columns
#Apply One-Hot Encoding to Categorical Features**
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred = rf_model.predict(X_test)

# Accuracy Score
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Neurl Networks model
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Convert labels to numpy arrays
y_train, y_test = np.array(y_train), np.array(y_test)
```

```python
from tensorflow.keras.regularizers import l1
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import AdamW

model = Sequential([
    Dense(128, activation='relu', kernel_regularizer=l1(0.01), input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(96, activation='relu', kernel_regularizer=l1(0.01)),  # 🔹 New Layer
    Dropout(0.3),
    Dense(64, activation='relu', kernel_regularizer=l1(0.01)),
    Dropout(0.3),
    Dense(48, activation='relu', kernel_regularizer=l1(0.01)),  # 🔹 New Layer
    Dense(32, activation='relu', kernel_regularizer=l1(0.01)),
    Dense(1)  # ⚠️ No sigmoid activation (logits output)
])


# Compile Model with `from_logits=True`
model.compile(optimizer=AdamW(learning_rate=0.001), loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")

# Predictions
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```
